# Calculate PI in an Approximization

In this notebook I try to calculate PI using a random drop example.

```python
from random import random
from matplotlib import pyplot as plt
from numpy import array

N = 10000
```

```python
drops = []
for iteration in range(N):
    x, y = random(), random()
    r = (x ** 2 + y ** 2) ** 0.5
    drops.append(list((x, y, r)))
```

```python
insiders = array([[x, y] for x, y, r in drops if r <= 1])
outsiders = array([[x, y] for x, y, r in drops if r > 1])
drops = array(drops)
print("Appoximated size of PI =", 4 * len(insiders) / len(drops))
```

```python
plt.scatter(x=insiders[:,0], y=insiders[:,1], color="grey", label="insiders")
plt.scatter(x=outsiders[:,0], y=outsiders[:,1], color="red", label="outsiders")
plt.legend(loc=3)
plt.show()
```
