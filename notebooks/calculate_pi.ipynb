```python
from random import random


N = 100000
```

```python
drops = []
for iteration in range(N):
    x, y = random(), random()
    r = (x ** 2 + y ** 2) ** 0.5
    drops.append(r)
```

```python
outsiders = len([val for val in drops if val > 1])
insiders = len([val for val in drops if val <= 1])
print(outsiders/insiders)
```
