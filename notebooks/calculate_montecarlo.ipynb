# Montecarlo Simulation

In this Python file, I am doing a simple Montecarlo Simulation to test the concept and try out new coding styles.

```python
from numpy.random import normal
from numpy import cumsum, array, vstack
from matplotlib import pyplot as plt
import seaborn as sns

N_STEPS = 1000
N_WALKS = 1000
START = 100
```

```python
random_walks = []

for i in range(N_WALKS):
    deltas = normal(0, 1, N_STEPS)
    walk = START + cumsum(deltas)
    random_walks.append(walk)

random_walks = array(random_walks)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharey=True)
for walk in random_walks:
    ax[0].plot(walk)
ax[1].hist(random_walks[-1,:], orientation='horizontal', bins=30)
plt.show()
```
