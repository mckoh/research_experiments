# Simple SAX-Experiement

```python
from saxpy.sax import ts_to_string
from saxpy.znorm import znorm
from saxpy.alphabet import cuts_for_asize
```

```python
timeseries = [1.2, 9.5, 1.3, 1.2, 1.5, 1.1]
```

```python
char_timeseries = ts_to_string(znorm(timeseries), cuts_for_asize(5))
```

```python
print(char_timeseries)
```
